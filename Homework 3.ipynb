```python
h = pd.read_csv('data/NYC_Health_and_Hospitals_Corporation_Facilities.csv')
```

```python
h.info()
```

```python
health_map = h[(h.Latitude.isna() == False) & (h.Longitude.isna() == False)]
```

```python
print (health_map.Latitude)
```

```python
m = folium.Map(location = [health_map.Latitude[1], health_map.Longitude[1]], tiles='Mapbox Bright',) 
m
```

```python
for index, row in health_map.iterrows():
    folium.CircleMarker(location=(row["Latitude"],
                                  row["Longitude"]),
                        radius = 1.0,
                        color='red',
                        fill=True).add_to(m)

folium.Marker([40.739173, -73.976862], popup='Bellevue Hospital Center').add_to(m)  
folium.Marker([40.784557, -73.943780], popup='Metropolitan Hospital Center').add_to(m)
folium.Marker([40.717250, -73.803433], popup='Queens Hospital Center').add_to(m)
folium.Marker([40.880462, -73.881640], popup='North Central Bronx Hospital').add_to(m)
folium.Marker([40.655762, -73.944580], popup='Kings County Hospital Center').add_to(m)
folium.Marker([40.586552, -73.966168], popup='Coney Island Hospital').add_to(m)
```
