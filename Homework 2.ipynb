```python
#Dataset for ratings of Japanese Hotels 
```

```python
import pandas as pd
```

```python
japan = pd.read_csv('data/Hostel.csv')
```

```python
japan
```

```python
japan.info()
```

```python
japan.groupby('City')['valueformoney'].count()
```

```python
city = japan.groupby('City')['valueformoney'].count()
```

```python
city
```

```python
type(city)
```

```python
city.index
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.bar(city.index, city)
```

```python
japan.groupby('rating.band')['valueformoney'].count()
```

```python
rating = japan.groupby('rating.band')['valueformoney'].count()
```

```python
plt.bar(rating, rating.index)
```

```python
japan.groupby('atmosphere')['valueformoney'].count().nlargest(5)
```

```python
japan.groupby('atmosphere')['valueformoney'].count().nsmallest(5)
```

```python
u5 = japan[japan['valueformoney'] <= 5]
```

```python
u5.head()
```

```python
u5.nlargest(10, 'summary.score')
```

```python
japan.groupby('City')['price.from'].mean().sort_values(ascending = False).plot(kind = 'bar')
```
